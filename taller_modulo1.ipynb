### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

# Promedio
suma = 0

for esperanza_de_vida in esperanzas_de_vida:
    suma = suma + esperanza_de_vida

promedio = suma / len(esperanzas_de_vida) #Len: longitud de la lista, número de datos de la lista.
print ("El promedio de la lista es: ", promedio)

# Máximo

maximo = esperanzas_de_vida[0] #Se asume la primera posición de la lista como el valor máximo, de forma inicial.

for esperanza_de_vida in esperanzas_de_vida:
    if esperanza_de_vida > maximo: #Si el valor de cada posición de la lista es mayor a maximo entonces se le asigna el valor a la variable maximo.
        maximo = esperanza_de_vida 

print("El valor máximo de la lista es: ", maximo)

# Mínimo

minimo = esperanzas_de_vida[0] #Se asume la primera posición de la lista como el valor mínimo, de forma inicial.

for esperanza_de_vida in esperanzas_de_vida:
    if esperanza_de_vida < minimo:
        minimo = esperanza_de_vida

print("El valor mínimo de la lista es: ", minimo)

````

### 1.3 — Funciones

Las funciones nos permiten reutilizar lógica.

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí

def clasificar_esperanza_vida(años): #Se establece la función mediante def, con el parámetro años
    if años >=75:
        return("Alta")
    elif 60 <= años <=74.9:
        return("Media")
    else:
        return("Baja")

print(clasificar_esperanza_vida(25)) #Llamando la función clasificar_esperanza_vida
print(clasificar_esperanza_vida(60))
print(clasificar_esperanza_vida(70))
print(clasificar_esperanza_vida(86))
print(clasificar_esperanza_vida(12))
````

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí
#1. Esperanza de vida promedio por region

salud.groupby("region")["esperanza_vida"].mean()

````

````python
#2. Gasto en salud promedio por nivel_ingresos

salud.groupby("nivel_ingresos")["gasto_salud_usd"].mean()

#De acuerdo a los datos obtenidos, podemos observar que el nivel de ingresos es directamente proporcional al gasto en salud promedio
#A mayor nivel de ingresos mayor gasto en salud.
````

````python
#3. Número de países por región
salud.groupby("region")["pais"].count()
````

````python
#4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?
promedio = salud.groupby("region")["esperanza_vida"].mean() #Esperanza de vida promedio por región
print(promedio)

#Mayor esperanza de vida promedio:
promedio.idxmax()
print("La región con mayor esperanza de vida promedio es ", promedio.idxmax())

#Menor esperanza de vida promedio:
promedio.idxmin()
print("La región con menor esperanza de vida promedio es ", promedio.idxmin())
````
